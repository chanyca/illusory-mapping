```python
import matplotlib.pyplot as plt
from matplotlib.gridspec import GridSpec
import importlib

# Custom imports
import plotting
importlib.reload(plotting)
from plotting import Plots
plotting = Plots()
```

```python
fig = plt.figure(figsize=(12, 3), constrained_layout=True, dpi=200)
gs = GridSpec(nrows=2, ncols=8, figure=fig)
ax = None

# Polar plot to show stimulus location
ax_A = fig.add_subplot(gs[0:2, 0:2], projection='polar')
ax=ax_A
plotting.polar_schematics(ax)

# F0B0
ax_B = fig.add_subplot(gs[0, 2])
ax=ax_B
plotting.condition_schematics(ax, n_flash=0, n_beep=0)
ax.set_title('F0B0')
ax.set_axis_off

# F1B0
ax_C = fig.add_subplot(gs[0, 3])
ax=ax_C
plotting.condition_schematics(ax, n_flash=1, n_beep=0)
ax.set_title('F1B0')

# F1B1
ax_D = fig.add_subplot(gs[1, 2])
ax=ax_D
plotting.condition_schematics(ax, n_flash=1, n_beep=1)
ax.set_title('F1B1')

# F1B2
ax_E = fig.add_subplot(gs[1, 3])
ax=ax_E
plotting.condition_schematics(ax, n_flash=1, n_beep=2)
ax.set_title('F1B2')

# Add labels
top_y, bottom_y = 0.98, 0.52
col1, col2, col3 = 0.05, 0.25, 0.38
fontsize = 11
plt.gcf().text(col1, top_y, "A", fontweight='bold', fontsize=fontsize)
plt.gcf().text(col2, top_y, "B", fontweight='bold', fontsize=fontsize)
plt.gcf().text(col3, top_y, "C", fontweight='bold', fontsize=fontsize)
plt.gcf().text(col2, bottom_y, "D", fontweight='bold', fontsize=fontsize)
plt.gcf().text(col3, bottom_y, "E", fontweight='bold', fontsize=fontsize)

# Add legend
handles, labels = ax.get_legend_handles_labels()
fig.legend(handles, labels, ncol=2, loc="center left", bbox_to_anchor=(0.3, -0.05))


fig.subplots_adjust(bottom=0.1)
plt.show()
```
